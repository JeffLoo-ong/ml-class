```python
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"
import numpy as np

from skimage import io
from skimage import data
from skimage.measure import block_reduce
```

```python
image = io.imread('dog.jpg', as_gray=True)
new_image = block_reduce(image, block_size=(2, 2), func=np.average)
new_image = (new_image - np.min(new_image)) / np.ptp(new_image)
io.imshow(new_image)
```
